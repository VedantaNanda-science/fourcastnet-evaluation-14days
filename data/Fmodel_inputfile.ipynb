### 2. Variable on pressure level : u-wind

### 3. Variable on pressure level : v-wind

```python
ds6file_path = "/content/drive/My Drive/Fourcastnet/era5_pressure_vwind_fmodel.nc"

# Load the NetCDF file
ds6 = xr.open_dataset(ds6file_path)
ds6
```

### Merging Datasets

```python
merged_ds_2 = xr.merge([ds4, ds5, ds6])
merged_ds_2
```

### Merging merged_ds and merged_ds_2
merged_ds contains variables: sst, mslp, tp (without pressure level)

merged_ds_2 contains 4D variables ( t, u, v, z) with shape: [time, pressure level, lat, lon]

We will start by first aligning dimensions and expanding merged_ds to add a dummy level dimension, so it can merge cleanly.

```python
# Match the level dimension from merged_ds_2
pressure_levels = merged_ds_2['pressure_level']

# Expand merged_ds with level dimension (broadcast surface vars across levels)
expanded_ds = merged_ds.expand_dims({'pressure_level': pressure_levels})

# Merge datasets
combined_ds = xr.merge([expanded_ds, merged_ds_2])
combined_ds
```

### Saving the Final Dataset to Drive

```python
combined_ds.to_netcdf("/content/drive/My Drive/Fourcastnet/forecastnet_input_final.nc")
```
